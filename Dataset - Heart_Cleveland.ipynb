# Paso 2. Preprocesamiento los Datos

### Eliminar valores atípicos en chol, trestbps y oldpeak usando el método del rango intercuartil (IQR).

```python
import numpy as np

# Función para eliminar valores atípicos usando el rango intercuartil (IQR)
def remove_outliers(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)]

# Aplicar eliminación de valores atípicos en `chol`, `trestbps` y `oldpeak`
df_clean = remove_outliers(df, ["chol", "trestbps", "oldpeak"])

# Mostrar cantidad de filas eliminadas
print(f"Filas eliminadas por valores atípicos: {df.shape[0] - df_clean.shape[0]}")
```

### Normalizar variables continuas (age, trestbps, chol, thalach, oldpeak) para mejorar el rendimiento del modelo.

```python
from sklearn.preprocessing import MinMaxScaler

# Columnas a normalizar
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

# Aplicar MinMaxScaler
scaler = MinMaxScaler()
df_clean.loc[:, num_cols] = scaler.fit_transform(df_clean[num_cols])

# Ver primeras filas después de la normalización
df_clean.head()
```

### Resultados

#### Valores Atípicos Eliminados

Se eliminaron 18 filas que contenían valores atípicos en chol, trestbps y oldpeak.

Esto ayuda a mejorar la estabilidad del modelo al evitar sesgos extremos.

#### Normalización Exitosa

Las columnas age, trestbps, chol, thalach, oldpeak han sido escaladas entre 0 y 1.

Ahora, las variables tienen una escala uniforme, lo que beneficia a modelos como la Regresión Logística y Árboles de Decisión.

# Paso 3. Selección de características más relevantes

#### con SelectKBest (Chi-cuadrado)

```python
from sklearn.feature_selection import SelectKBest, chi2

# Separar variables predictoras (X) y objetivo (y)
X = df_clean.drop(columns=["condition"])
y = df_clean["condition"]

# Aplicar SelectKBest con chi-cuadrado
selector = SelectKBest(score_func=chi2, k=6)  # Elegimos las 6 mejores características
X_new = selector.fit_transform(X, y)

# Mostrar las mejores características
selected_features = X.columns[selector.get_support()]
print("Características seleccionadas:", selected_features)
```

### Análisis de la Selección de Características

Las 6 características más relevantes para predecir condition son:

**cp (tipo de dolor en el pecho):** Se esperaba que fuera relevante, ya que en el EDA mostró fuerte correlación con la enfermedad.

**exang (angina inducida por el ejercicio)**: Indica si la persona experimenta dolor en el pecho durante el ejercicio.

**oldpeak (depresión del ST después del ejercicio):** Fuertemente correlacionada con la enfermedad cardíaca.

**slope (pendiente del segmento ST en el ECG):** Relacionada con cambios en la actividad del corazón.

**ca (cantidad de vasos coloreados en la angiografía)**: Relacionado con el flujo sanguíneo en el corazón.

**thal (tipo de talasemia):** Se asocia con anomalías cardíacas.



# Paso 4. División del dataset en Train y Test

```python
from sklearn.model_selection import train_test_split

# Usamos solo las características seleccionadas
X_selected = df_clean[selected_features]
y = df_clean["condition"]

# Dividir en 80% entrenamiento y 20% prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)

# Mostrar tamaños de los conjuntos
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape[0]} filas")
print(f"Tamaño del conjunto de prueba: {X_test.shape[0]} filas")
```

### Análisis de la División del Dataset

80% Entrenamiento (223 filas) → Para que el modelo aprenda correctamente.

20% Prueba (56 filas) → Para evaluar el rendimiento en datos nuevos.

stratify=y garantiza que ambas clases (enfermos y sanos) estén equilibradas en los conjuntos.

# Paso 5. Entrenar el modelo

### 1. Entrenamiento del Modelo - Regresión Lineal

```python
from sklearn.linear_model import LinearRegression

# Crear y entrenar el modelo
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_linear = linear_model.predict(X_test)

# Mostrar los primeros valores predichos
print("Primeras predicciones de Regresión Lineal:", y_pred_linear[:10])
```

### Análisis de las Predicciones de Regresión Lineal

##### Valores de salida:

La regresión lineal genera valores continuos en el rango 0 a 1, pero nuestro problema es de clasificación (0 o 1).

Para interpretar estos resultados, podemos redondear los valores:

≥ 0.5 → Clase 1 (enfermedad cardíaca presente)

< 0.5 → Clase 0 (sin enfermedad cardíaca)



### Evaluación de Regresión Lineal



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convertir predicciones continuas a 0 y 1
y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred_linear]

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred_binary)
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

# Mostrar resultados
print(f"Precisión: {accuracy:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

### Análisis del Desempeño de Regresión Lineal

Precisión: 83.93% → Buen porcentaje de predicciones correctas en general.

Recall: 80.00% → El modelo identifica correctamente el 80% de los pacientes con enfermedad cardíaca.

F1-Score: 81.63% → Buen balance entre precisión y recall.

"La Regresión Lineal funciona bien como un modelo base, pero puede no ser ideal para clasificación."



### 2. Entrenamiento del Modelo - Regresión Logística

```python
from sklearn.linear_model import LogisticRegression

# Crear y entrenar el modelo de Regresión Logística
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_logistic = logistic_model.predict(X_test)

# Evaluación del modelo
accuracy_log = accuracy_score(y_test, y_pred_logistic)
precision_log = precision_score(y_test, y_pred_logistic)
recall_log = recall_score(y_test, y_pred_logistic)
f1_log = f1_score(y_test, y_pred_logistic)

# Mostrar resultados
print(f"Precisión: {accuracy_log:.4f}")
print(f"Recall: {recall_log:.4f}")
print(f"F1-Score: {f1_log:.4f}")
```

### Análisis

Regresión Lineal tiene una precisión ligeramente mejor (83.93% vs. 82.14%), pero la diferencia no es significativa.

Recall es idéntico (80%), lo que significa que ambos modelos identifican la misma cantidad de casos positivos correctamente.

F1-Score es similar, pero la Regresión Lineal obtiene un ligero mejor equilibrio.

#### Conclusión:

Regresión Logística es más adecuada porque está diseñada para clasificación binaria.

Regresión Lineal puede ser útil como referencia, pero no es la mejor opción para este problema.



### 3. Entrenamiento del Modelo - Árbol de Decisión



```python
from sklearn.tree import DecisionTreeClassifier

# Crear y entrenar el modelo de Árbol de Decisión
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_tree = tree_model.predict(X_test)

# Evaluación del modelo
accuracy_tree = accuracy_score(y_test, y_pred_tree)
precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

# Mostrar resultados
print(f"Precisión: {accuracy_tree:.4f}")
print(f"Recall: {recall_tree:.4f}")
print(f"F1-Score: {f1_tree:.4f}")
```

### Análisis

El Árbol de Decisión tiene el peor desempeño de los tres modelos.

La precisión bajó a 71.43%, lo que significa que está cometiendo más errores en las predicciones.

Recall cayó a 72%, lo que indica que detecta menos casos positivos correctamente.

F1-Score bajó a 69.23%, mostrando que el modelo tiene un menor balance entre precisión y recall.

#### Conclusión:

El Árbol de Decisión no es tan efectivo en este dataset con su configuración actual.

Puede mejorar ajustando hiperparámetros como profundidad del árbol o usando poda.

Regresión Logística sigue siendo la mejor opción por ahora debido a su estabilidad y buen balance.

#### Optimización del Árbol de Decisión con GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Definir el modelo
tree_model = DecisionTreeClassifier(random_state=42)

# Definir los hiperparámetros a probar
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Aplicar GridSearchCV
grid_search = GridSearchCV(tree_model, param_grid, cv=5, scoring='f1')
grid_search.fit(X_train, y_train)

# Mejor modelo encontrado
best_tree = grid_search.best_estimator_

# Evaluar en el conjunto de prueba
y_pred_best_tree = best_tree.predict(X_test)
accuracy_best_tree = accuracy_score(y_test, y_pred_best_tree)
precision_best_tree = precision_score(y_test, y_pred_best_tree)
recall_best_tree = recall_score(y_test, y_pred_best_tree)
f1_best_tree = f1_score(y_test, y_pred_best_tree)

# Mostrar resultados optimizados
print("Mejores hiperparámetros:", grid_search.best_params_)
print(f"Precisión Optimizada: {accuracy_best_tree:.4f}")
print(f"Recall Optimizado: {recall_best_tree:.4f}")
print(f"F1-Score Optimizado: {f1_best_tree:.4f}")
```

### Análisis

Optimización exitosa, el Árbol de Decisión mejoró en todas las métricas.

La precisión subió de 71.43% a 80.36%, acercándose a los modelos de regresión.

El F1-Score mejoró de 69.23% a 76.60%, lo que indica un mejor equilibrio.

Aún no supera a Regresión Lineal o Logística, pero es una opción válida con reglas interpretables.

#### Conclusión:

La Regresión Lineal sigue siendo el mejor modelo, pero la Regresión Logística es la mejor opción para clasificación.

El Árbol de Decisión optimizado es más competitivo y ofrece interpretabilidad.
